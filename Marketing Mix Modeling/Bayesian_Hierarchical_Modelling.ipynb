To allow the model to learn that "No spend == No contribution", we must fill in all NaN values with 0.

These values cannot be dropped.

```python
for col in ["google_spend", "meta_spend", "tiktok_spend"]:
    df[col] = df[col].fillna(0.0)
```

Log transforms and scaling continuous features

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

scale_cols = ["google_spend", "meta_spend", "tiktok_spend"]

for col in scale_cols:
    df[f"log_{col}"] = np.log1p(df[col])

scaler = StandardScaler()
df[[f"z_{c}" for c in scale_cols]] = scaler.fit_transform(
    df[[f"log_{c}" for c in scale_cols]]
)
```

```python
control_cols = [
    "ORGANIC_SEARCH_CLICKS",
    "BRANDED_SEARCH_CLICKS",
    "DIRECT_CLICKS"
]

for col in control_cols:
    df[col] = df[col].fillna(0.0)
    df[f"log_{col}"] = np.log1p(df[col])

df[[f"z_{c}" for c in control_cols]] = scaler.fit_transform(
    df[[f"log_{c}" for c in control_cols]]
)
```

```python
df["org_idx"] = df["ORGANISATION_ID"].astype("category").cat.codes
df["campaign_idx"] = df["campaign_category"].astype("category").cat.codes
```

```python
org_lookup = dict(enumerate(df["ORGANISATION_ID"].astype("category").cat.categories))
campaign_lookup = dict(enumerate(df["campaign_category"].astype("category").cat.categories))
```

This is the final modeling matrix that the bayesian model will see

```python
X_media = df[[
    "z_google_spend",
    "z_meta_spend",
    "z_tiktok_spend"
]].values

X_controls = df[[
    "z_ORGANIC_SEARCH_CLICKS",
    "z_BRANDED_SEARCH_CLICKS",
    "z_DIRECT_CLICKS"
]].values

y = df["log_purchases"].values

org_idx = df["org_idx"].values
campaign_idx = df["campaign_idx"].values
```

# Working with PyMC to create Bayesian Model

```python
import pymc as pm
import numpy as np
import arviz as az
```

```python
# Development subsample
np.random.seed(42)
sample_idx = np.random.choice(len(y), size=25000, replace=False)

y_dev = y[sample_idx]
X_media_dev = X_media[sample_idx]
X_controls_dev = X_controls[sample_idx]
org_idx_dev = org_idx[sample_idx]
campaign_idx_dev = campaign_idx[sample_idx]
```

```python
coords = {
    "obs_id": np.arange(len(y_dev)),
    "org": np.unique(org_idx),
    "campaign": np.unique(campaign_idx),
    "media_channel": ["google", "meta", "tiktok"],
    "control": ["organic", "branded", "direct"]
}
```

```python
print(X_media_dev.shape)      # (N, 3)
print(X_controls_dev.shape)   # (N, 3)
print(y_dev.shape)            # (N,)
print(org_idx_dev.max() + 1)  # number of orgs
print(campaign_idx_dev.max() + 1)  # number of campaign categories
```

```python
with pm.Model(coords=coords) as hierarchical_mmm:

    # -------------------------
    # DATA
    # -------------------------
    y_obs = pm.Data("y_obs", y_dev, dims="obs_id")
    org = pm.Data("org_idx", org_idx_dev, dims="obs_id")
    campaign = pm.Data("campaign_idx", campaign_idx_dev, dims="obs_id")

    X_media_data = pm.Data("X_media", X_media_dev, dims=("obs_id", "media_channel"))
    X_controls_data = pm.Data("X_controls", X_controls_dev, dims=("obs_id", "control"))

    # -------------------------
    # GLOBAL PRIORS
    # -------------------------
    alpha_global = pm.Normal("alpha_global", mu=0.0, sigma=1.0)

    beta_media_global = pm.Normal(
        "beta_media_global",
        mu=0.0,
        sigma=0.5,
        dims="media_channel"
    )

    # -------------------------
    # HYPERPRIORS (SHRINKAGE)
    # -------------------------
    sigma_org = pm.HalfNormal("sigma_org", sigma=0.3)

    sigma_media = pm.HalfNormal(
        "sigma_media",
        sigma=0.3,
        dims="media_channel"
    )

    # -------------------------
    # RANDOM EFFECTS
    # -------------------------
    alpha_org = pm.Normal(
        "alpha_org",
        mu=alpha_global,
        sigma=sigma_org,
        dims="org"
    )

    beta_media_campaign = pm.Normal(
        "beta_media_campaign",
        mu=beta_media_global,
        sigma=sigma_media,
        dims=("campaign", "media_channel")
    )

    # -------------------------
    # CONTROL EFFECTS (POOLED)
    # -------------------------
    beta_controls = pm.Normal(
        "beta_controls",
        mu=0.0,
        sigma=0.5,
        dims="control"
    )

    # -------------------------
    # LINEAR PREDICTOR
    # -------------------------
    mu = (
        alpha_org[org]
        + pm.math.sum(beta_media_campaign[campaign] * X_media_data, axis=1)
        + pm.math.sum(beta_controls * X_controls_data, axis=1)
    )

    # -------------------------
    # LIKELIHOOD
    # -------------------------
    sigma = pm.HalfNormal("sigma", sigma=1.0)

    pm.Normal(
        "likelihood",
        mu=mu,
        sigma=sigma,
        observed=y_obs,
        dims="obs_id"
    )
```

```python
with hierarchical_mmm:
    trace = pm.sample(
        draws=1000,
        tune=1000,
        chains=2,
        target_accept=0.97,
        return_inferencedata=True
    )
```

# Step 5 - Summaries and credible intervals and Elasticities

```python
az.summary(
    trace,
    var_names=["beta_media_global", "sigma_media", "sigma"],
    round_to=2
)
```

```python
global_media_effects = az.summary(
    trace,
    var_names=["beta_media_global"],
    hdi_prob=0.95
)

global_media_effects
```

```python
campaign_media_effects = az.summary(
    trace,
    var_names=["beta_media_campaign"],
    hdi_prob=0.95
)

campaign_media_effects
```

```python
effects_df = campaign_media_effects.reset_index()
# Extract the campaign index (digit) and map it to the category name
effects_df["campaign"] = effects_df["index"].str.extract(r"\[(\d+),", expand=False).astype(int).map(campaign_lookup)

# Create a mapping from channel names (strings) to their integer indices
channel_name_to_int = {name: i for i, name in enumerate(coords["media_channel"])}

# Extract the channel name (word) and then map it to its integer index
effects_df["channel"] = effects_df["index"].str.extract(r",\s*(\w+)\]", expand=False)
effects_df["channel"] = effects_df["channel"].map(channel_name_to_int).astype(int)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.pointplot(
    data=effects_df,
    x="channel",
    y="mean",
    hue="campaign",
    dodge=True
)
plt.axhline(0, color="black", linestyle="--")
plt.title("Campaign-level Channel Effects with Hierarchical Shrinkage")
plt.ylabel("Posterior Mean Effect")
plt.show()
```

```python
elasticities = {}

for channel, col in zip(
    ["Google", "Meta", "TikTok"],
    ["log_google_spend", "log_meta_spend", "log_tiktok_spend"]
):
    elasticity = (
        global_media_effects.loc[f"beta_media_global[{channel.lower()}]", "mean"]
        * df[col].std()
        / df["log_purchases"].std()
    )
    elasticities[channel] = elasticity

elasticities
```
